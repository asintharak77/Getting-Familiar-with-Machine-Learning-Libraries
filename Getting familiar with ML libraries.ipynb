### 8. Apply MinMaxScaler() and StandardScaler() to the following features: calories, protien, fat, sodium, fiber, carbo, sugars.

### MinMaxScaler()

Normalization scales each input variable separately to the range 0-1, which is the range for floating-point values where we have the most precision

A value is normalized as follows:

**y = (x – min) / (max – min)**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

cereals[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo','sugars']] = scaler.fit_transform(cereals[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo','sugars']])
```

```python
cereals
```

```python
features = ['calories', 'protein', 'fat', 'sodium']
for i in features:
    cereals[[i]].hist()
```

### StandardScaler()

Standardizing a dataset involves rescaling the distribution of values so that the mean of observed values is 0 and the standard deviation is 1.

The values are standardized as

**y = (x – mean) / standard_deviation**

Where the mean is calculated as:

**mean = sum(x) / count(x)**

And the standard_deviation is calculated as:

**standard_deviation = sqrt( sum( (x – mean)^2 ) / count(x))**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cereals[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo','sugars']] = scaler.fit_transform(cereals[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo','sugars']])
```

```python
cereals
```

```python
for i in features:
    cereals[[i]].hist()
```

### 9. Does the standard or min-max scaling make a difference in value distribution? Support your answers with some visualisations on the above dataset.

Yes, the standard or min-max scaling make a difference in value distribution. Because MinMaxScaler transform features by scaling each feature to a given range. But StandardScaler removes the mean and scales each feature/variable to unit variance.


We can understand this difference from the above plotted graphs.

**MinMaxScaler-Graphs**

- The data lies between 0 and 1

**StandardScaler-Graphs**

- Rescales the distribution of values such that the mean of observation is 0 and their standard deviation is 1
- It may not always belong to the range 0 to 1
